```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
```

```python
gambar_original = imageio.imread("segmen.jpg")

plt.imshow(gambar_original)
plt.title("Citra original")
plt.show()
```

```python
img = imageio.imread("segmen.jpg")

img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
```

```python
img_inversi = np.zeros(img.shape, dtype=np.uint8)
```

```python
def inversi_grayscale(nilai):
  for y in range(0, img_height):
    for x in range(0, img_width):
      red = img[y][x][0]
      green = img[y][x][1]
      blue = img[y][x][2]
      gray = (int(red) + int(green) + int(blue)) / 3
      gray = nilai - gray
      img_inversi[y][x] = (gray, gray, gray)
```

```python
def inversi_rgb(nilai):
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            red = nilai - red
            green = img[y][x][1]
            green = nilai - green
            blue = img[y][x][2]
            blue = nilai - blue
            img_inversi[y][x] = (red, green, blue)
```

```python
inversi_grayscale(255)
plt.imshow(img_inversi)
plt.title("inversi grayscale")
plt.show()

```

```python
inversi_rgb(255)
plt.imshow(img_inversi)
plt.title("inversi rgb")
plt.show()
```
